### 7) Show all the courses of Graphic Design where price is below 100 !

```python
data.head(2)
```

```python
# Method - 1

#using filter method, 2 times

# 1st for subject = Graphic Design
df_2 = data[data['subject'] == 'Graphic Design']

# Again applying the filter for price col < 100

# applying condition when less than 100 and also when it is free
# the price col is of string data type
(df_2['price'] < '100') | (df_2['price'] == 'Free')

df_2[(df_2['price'] < '100') | (df_2['price'] == 'Free')]
```

```python
# Method - 2

# condition
(data['subject'] == 'Graphic Design') & ((data['price'] < '100') | (data['price'] == 'Free'))

# dataframe for above condition
data[(data['subject'] == 'Graphic Design') & ((data['price'] < '100') | (data['price'] == 'Free'))]
```

### 9) What are courses that published in year 2015

```python
# checking the data types
data.dtypes
```

```python
# published timestamp have object data type
# converting it to datetime
pd.to_datetime(data['published_timestamp'])

# and assigning to itself column
data['published_timestamp'] = pd.to_datetime(data['published_timestamp'])

# printing data types
data.dtypes
```

```python
# now extracting the year from it
data['published_timestamp'].dt.year

# comparing with 2015 will give boolean result
data['published_timestamp'].dt.year == 2015

# getting dataframe
data[data['published_timestamp'].dt.year == 2015]
```

### 10) What are the maximum number of subscriber for each level of courses ?

```python
data.head(2)
```

```python
# groupby course title
data.groupby('level')

# selecting 'num_subscribers' column
data.groupby('level')['num_subscribers']

# getting the maximum out of it
data.groupby('level')['num_subscribers'].max()
```

```python
# method 2
data.groupby('level').max()
```
