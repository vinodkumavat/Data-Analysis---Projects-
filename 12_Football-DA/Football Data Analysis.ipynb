### 4) Draw pie chart for penalties missed vs scored

```python
data.head(2)
```

```python
# penalty missed = Total penalty attempted - successfull penalty
penalty_missed = data['Penalty_Attempted'].sum() - data['Penalty_Goals'].sum()
penalty_scored = data['Penalty_Goals'].sum()

# creating list data for pie chart
list_data = [penalty_scored, penalty_missed]

# plotting 
labels_ = ['Penalty-Missed', 'Penalty-Scored']
color_ = sns.color_palette('Set2')

plt.pie(list_data, labels = labels_, colors = color_, autopct = '%.0f%%')
plt.show()
```

### 7) Most of the players are from which nations

```python
data.head(2)
```

```python
# groupby Nationality
data.groupby('Nationality')

# getting the size
data.groupby('Nationality').size()

# sorting them 
data.groupby('Nationality').size().sort_values(ascending = False)

# getting top 10
data.groupby('Nationality').size().sort_values(ascending = False).head(10)

# plotting them
data.groupby('Nationality').size().sort_values(ascending = False).head(10).plot(kind = 'bar', color = sns.color_palette('magma'))
```
