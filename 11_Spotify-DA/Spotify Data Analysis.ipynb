### 6) Change duration column from "ms" to "sec"

```python
# selecting the column
data['duration_ms']

# dividing each elements by 1000 to get the values in seconds
data['duration_ms'].apply(lambda x : round(x/1000, 2))

# assining it to new column
data['duration_ms'] = data['duration_ms'].apply(lambda x : round(x/1000, 2))

# change the name of column from duration_ms to duration_sec
data = data.rename(columns = {'duration_ms':'duration_sec'})

# print data
data.head(3)
```

.
### 7) Create regression plot between energy and loudness

```python
sample_df = data.sample(int(0.004*len(data)))
sample_df
```

```python
plt.figure()
sns.regplot(data = sample_df, y = 'loudness', x = 'energy', color = 'g').set(title = 'Loudness vs Energy correlation')
plt.show
```
